```python
x_data = pd.read_csv('/kaggle/input/2021-ml-p4/train_desc_pca.csv')
y_data = pd.read_csv('/kaggle/input/2021-ml-p4/train_label.csv')
test_data = pd.read_csv('/kaggle/input/2021-ml-p4/test_desc_pca.csv')
submit = pd.read_csv('/kaggle/input/2021-ml-p4/sample_submit.csv')

print(x_data.head())
print(y_data.head())
print(test_data.head())
```

```python
# label데이터 id지우기
y_data = y_data['Category']
print(y_data.head())
print(y_data.unique())
```

```python
# train data를 validation으로 나누기
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.3, random_state=123, stratify = y_data)

print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
# train
from sklearn.linear_model import LogisticRegression

logist = LogisticRegression(C=200.0)

l2_100 = LogisticRegression(penalty = 'l2', C= 100.0, solver = 'saga', random_state=1, max_iter=500)
l2_10 = LogisticRegression(penalty = 'l2', C= 10.0, solver = 'saga', random_state=1, max_iter=500)
l2_1 = LogisticRegression(penalty = 'l2', C= 1.0, solver = 'saga', random_state=1, max_iter=500)
l2_01 = LogisticRegression(penalty = 'l2', C= 0.1, solver = 'saga', random_state=1, max_iter=500)

l1_100 = LogisticRegression(penalty = 'l1', C= 100.0, solver = 'saga', random_state=1, max_iter=500)
l1_10 = LogisticRegression(penalty = 'l1', C= 10.0, solver = 'saga', random_state=1, max_iter=500)
l1_1 = LogisticRegression(penalty = 'l1', C= 1.0, solver = 'saga', random_state=1, max_iter=5000)
l1_01 = LogisticRegression(penalty = 'l1', C= 0.1, solver = 'saga', random_state=1, max_iter=500)
```

```python
logist.fit(x_train, y_train)
print(logist.score(x_val, y_val))

l2_100.fit(x_train, y_train)
print(l2_100.score(x_val, y_val))

l2_10.fit(x_train, y_train)
print(l2_10.score(x_val, y_val))

l2_1.fit(x_train, y_train)
print(l2_1.score(x_val, y_val))

l2_01.fit(x_train, y_train)
print(l2_01.score(x_val, y_val))

l1_100.fit(x_train, y_train)
print(l1_100.score(x_val, y_val))

l1_10.fit(x_train, y_train)
print(l1_10.score(x_val, y_val))

l1_1.fit(x_train, y_train)
print(l1_1.score(x_val, y_val))

l1_01.fit(x_train, y_train)
print(l1_01.score(x_val, y_val))
```

```python
print(l2_100.intercept_)
print(l2_10.intercept_)
print(l2_1.intercept_)
print(l2_01.intercept_)

print(l1_100.intercept_)
print(l1_10.intercept_)
print(l1_1.intercept_)
print(l1_01.intercept_)
```

```python
print(l2_100.coef_)
print(l2_10.coef_)
print(l2_1.coef_)
print(l2_01.coef_)
```

```python
print(l1_100.coef_)
print(l1_10.coef_)
print(l1_1.coef_)
print(l1_01.coef_)
```
