```python
import pandas as pd

# train data load
train_data = pd.read_csv('/kaggle/input/2021-ml-p3/train.csv')
print(train_data)
```

```python
# test data load
test_data = pd.read_csv('/kaggle/input/2021-ml-p3/test.csv')
print(test_data)

# submit csv save
submit = pd.read_csv('/kaggle/input/2021-ml-p3/sample.csv')
```

```python
# train data를 x, y data로 나누기
x = train_data.drop('label', axis=1)
print(x.shape)

y_data = train_data['label']
print(y_data)
```

```python
# y data(label) 실수화
from sklearn.preprocessing import LabelEncoder
import numpy as np

classle = LabelEncoder() # 실수화
y = classle.fit_transform(train_data['label'])
print('species label : ', np.unique(y))
```

```python
# original species label name
yo = classle.inverse_transform(y)
print('species name : ', np.unique(yo))
```

### KNN(1) - train data를 train data와 validation data로 나누고 학습 진행

```python
# x data를 test와 validation으로 나누기
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1, stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# knn 학습
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5, p = 2)
knn.fit(X_train, y_train) # 모델 검증
```

```python
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```

### KNN(2) - train data를 표준화 해보기

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train) # 표준화 적용

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors = 5, p=2)
knn.fit(X_train_std, y_train)
```

```python
y_train_pred_std = knn.predict(X_train_std)
y_test_pred_std = knn.predict(X_test_std)

print(accuracy_score(y_train, y_train_pred_std))
print(accuracy_score(y_test, y_test_pred_std)) # 표준화를 적용하지 않은 것이 성능이 더 좋으므로 표준화 안한 것 선택!
```

### KNN(3) - train data를 validation으로 나누지 않은 경우

```python
knn = KNeighborsClassifier(n_neighbors = 5, p = 2)
knn.fit(x, y) # 모델 검증
```

```python
y_label_pred = knn.predict(test_data)
```

```python
y_label_pred_names = classle.inverse_transform(y_label_pred)
```

```python
submit['label']= y_label_pred_names
submit.to_csv('submit.csv', mode='w', header= True, index= False)
```
